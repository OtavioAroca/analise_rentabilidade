# **3) Anlise dos Dados**

De acordo com o enunciado, a quantidade de vendas permaneceu a mesma, no entanto, houve uma reduo da rentabilidade.

1) O lucro diminuiu com o mesmo investimento ou investimento maior;

Portanto, um dos deflatores  de fato o custo com entrega.

O custo elevado pode se dar por diversos fatores, como:

* Custos de transporte: Os custos associados ao transporte, como combustvel, manuteno de veculos, seguros e salrios dos motoristas, podem afetar diretamente o preo da entrega.

* Distncia e localizao: Quanto mais distante o destino da entrega e mais remoto o local, maior pode ser o custo do transporte devido a despesas adicionais de combustvel e tempo de viagem.
* Peso e volume da carga: O peso e o volume da carga a ser entregue podem influenciar o custo do transporte. Cargas pesadas ou volumosas podem exigir veculos maiores ou mais capacidade de carga, o que pode resultar em custos mais elevados.
* Urgncia da entrega: Entregas urgentes ou com prazos apertados podem exigir servios expressos ou prioritrios, que geralmente tm custos mais altos do que opes de entrega padro.
* Tipo de entrega: O mtodo de entrega escolhido tambm pode afetar o preo. Por exemplo, a entrega porta a porta pode ser mais cara do que a entrega em um ponto de coleta.
* Tempo e sazonalidade: A demanda por servios de entrega pode variar ao longo do tempo e em diferentes pocas do ano. Em perodos de alta demanda, como durante as festas de fim de ano, os preos das entregas podem aumentar devido  maior concorrncia e disponibilidade limitada de recursos.
* Regulamentaes e tarifas: Regulamentaes governamentais, tarifas de pedgio e outras taxas podem impactar os custos de entrega, especialmente em rotas especficas ou em determinadas reas geogrficas.

### **Mas como que a varivel Custo Entrega se relaciona com as outras?**

Outra prtica importante  entender a correlao entre as variveis. O mais comum de se utilizar para medir a correlao entre as variveis,  a correlao de Pearson. Contudo, precisa-se verificar se os dados atendem s premissas para se utilizar a correlao de Pearson.

Ser realizado um teste Kolmogorov-Smirnov (KS) para verificar se os dados seguem uma distribuio normal.

### **Teste de Hiptese - Os dados seguem uma distribuio normal?**

**Hiptese Nula (H0):** Os dados seguem uma distribuio normal.

**Hiptese Alternativa (HA):** Os dados no seguem uma distribuio normal.

**Nvel de Significncia:** 0.05

```python
# Realizando o teste de Kolmogorov-Smirnov
ks_statistic, ks_p_value = kstest(df['Custo Entrega'], 'norm')

print("Estatstica de teste KS:", ks_statistic)
print("Valor-p:", ks_p_value)
```

```python
# Realizando o teste de Kolmogorov-Smirnov
ks_statistic, ks_p_value = kstest(df['Custo Marketing'], 'norm')

print("Estatstica de teste KS:", ks_statistic)
print("Valor-p:", ks_p_value)
```

```python
# Realizando o teste de Kolmogorov-Smirnov
ks_statistic, ks_p_value = kstest(df['Custo dos Produtos'], 'norm')

print("Estatstica de teste KS:", ks_statistic)
print("Valor-p:", ks_p_value)
```

Com base nos resultados do teste de Kolmogorov-Smirnov, rejeitamos a hiptese nula de que os dados seguem a distribuio terica especificada (H0). O valor-p associado ao teste foi significativamente baixo (p < 0.05), indicando que h evidncias suficientes para concluir que os dados no seguem a distribuio normal.

Portanto, os dados podem ser considerados como no provenientes de uma distribuio normal. Esta concluso sugere que os dados podem ter uma distribuio diferente da normal, e pode ser necessrio explorar outras distribuies tericas ou realizar anlises adicionais para entender melhor a distribuio dos dados.

O teste de Pearson pressupe que os dados sigam uma distribuio normal e que haja linearidade entre as variveis, duas premissas que no foram atendidas. Portanto, no faz sentido prosseguir com esse teste.

O teste de correlao de Spearman  uma alternativa no paramtrica ao teste de Pearson. Ele no faz suposies sobre a distribuio dos dados e  menos sensvel a distribuies no normais. Portanto,  mais adequado quando os dados no seguem uma distribuio normal ou quando a relao entre as variveis no  linear.

### **Mapa de Calor (Heatmap)**

```python
correlacao = df.drop(columns='Categoria').corr(method='spearman')

plt.figure(figsize=(10,8))
sns.heatmap(correlacao, annot=True, fmt= '.3f', linewidths=.06)
plt.title('Mapa de Carlor da Correlao de Spearman')
```

### **Teste de Hiptese**

**Hiptese Nula (H0):** A correlao  = 0 

**Hiptese Alternativa (Ha):** A correlao  diferente de 0

**Nvel de Significncia:** 0.05

```python
corr, p_valor = spearmanr(df['Custo Entrega'], df['Custo Marketing'])

print("Correlao de Spearman:", corr)
print("Valor-p:", p_valor)
```

```python
corr, p_valor = spearmanr(df['Custo Entrega'], df['Custo dos Produtos'])

print("Correlao de Spearman:", corr)
print("Valor-p:", p_valor)
```

Como o p-valor foi encontrado abaixo do nvel de significncia (0.05), podemos rejeitar fortemente a hiptese nula de que no h correlao entre as variveis. Portanto, h evidncias estatsticas significativas para suportar a existncia de uma correlao entre as variveis, embora o coeficiente de correlao de Spearman por si s no fornea informaes sobre a direo ou magnitude dessa correlao.
